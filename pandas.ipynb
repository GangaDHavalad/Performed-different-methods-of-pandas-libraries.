```python
import sqlite3
conn=sqlite3.connect("DRF.db")
conn.execute("""CREATE TABLE IF NOT EXISTS DRF(
ID INTEGER PRINMARY KET,
NAME TEXT, 
AGE INTEGER,
MARKS INTEGER
)"""
            )
```

```python
DATA=[
    (1,"Shravya",3,97),
    (2,"Ganavi",4,87),
    (3,"Ram",22,68),
    (4,"Lakshman",21,56)
]
```

```python
conn.executemany('INSERT OR IGNORE INTO DRF VALUES(?,?,?,?)', DATA)
```

```python
import pandas as pd
df=pd.read_sql("SELECT * FROM DRF", conn)
df.head()
```
